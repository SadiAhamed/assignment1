```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
data=pd.read_excel('D:/activity/iris.xls')
data.head()
plt.figure(figsize=(10,10))
plt.subplot(4,1,1)
plt.hist(data["SL"],color="red",rwidth=0.9)
plt.title("SL")
plt.xticks()
plt.yticks()

plt.figure(figsize=(10,10))
plt.subplot(4,1,2)
plt.hist(data["SW"],color="blue",rwidth=0.9)
plt.title("SW")
plt.xticks()
plt.yticks()

plt.figure(figsize=(10,10))
plt.subplot(4,1,3)
plt.hist(data["PL"],color="violet",rwidth=0.9)
plt.title("PL")
plt.xticks()
plt.yticks()

plt.figure(figsize=(10,10))
plt.subplot(4,1,3)
plt.hist(data["PW"],color="black",rwidth=0.9)
plt.title("PW")
plt.xticks()
plt.yticks()
plt.tight_layout()

```

### b)PAIRPLOT

```python
sns.pairplot(data)
```
